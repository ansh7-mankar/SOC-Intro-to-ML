### 5. Loops
Python has two main loop types: `for` and `while`.
Use `break` to exit a loop early, `continue` to skip to the next iteration, and `else` on a loop to run code when it finishes *without* a `break`.

````python
# --- Basic for loop over a range ---
print("Counting up:")
for i in range(1, 6):
    print(f"  i = {i}")
````

````python
# --- for loop over a list ---
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"  I like {fruit}")
````

````python
# --- enumerate: get index AND value at the same time ---
students = ["Alice", "Bob", "Charlie"]
for index, name in enumerate(students):
    print(f"  Seat {index + 1}: {name}")
````

````python
# --- zip: loop over two lists together ---
names  = ["Alice", "Bob", "Charlie"]
scores = [88, 73, 95]
for name, score in zip(names, scores):
    print(f"  {name} scored {score}")
````

````python
# --- while loop ---
n = 1
while n <= 5:
    print(f"  n = {n}")
    n += 1
````

````python
# --- break: stop the loop early ---
print("Looking for 'banana':")
for fruit in ["apple", "mango", "banana", "cherry"]:
    if fruit == "banana":
        print(f"  Found it! Stopping.")
        break
    print(f"  Not {fruit}, keep looking...")
````

````python
# --- continue: skip one iteration ---
print("Skipping even numbers:")
for i in range(1, 8):
    if i % 2 == 0:
        continue        # jump back to the top of the loop
    print(f"  {i}")
````

````python
# --- for-else: else runs only if no break occurred ---
print("Search with for-else:")
target = 7
for num in [1, 3, 5, 9]:
    if num == target:
        print(f"  Found {target}!")
        break
else:
    print(f"  {target} was not in the list.")
````

````python
# --- Nested loops: multiplication table ---
print("3x3 multiplication table:")
for row in range(1, 4):
    for col in range(1, 4):
        print(f"  {row} x {col} = {row * col}")
````

````python
# --- List comprehension: a compact for loop that builds a list ---
squares = [x ** 2 for x in range(1, 6)]
print("Squares:", squares)

evens = [x for x in range(10) if x % 2 == 0]
print("Even numbers:", evens)
````

### 9. Classes and Objects
A **class** is a blueprint. An **object** is a specific instance built from that blueprint.

- `__init__` is the constructor — it runs automatically when you create a new object.
- `self` refers to the current object. Every method must take `self` as its first parameter.
- **Attributes** store data (`self.name`, `self.age`).
- **Methods** are functions that belong to the class.

````python
# --- Defining a simple class ---
class Dog:
    # Class attribute (shared by ALL Dog objects)
    species = "Canis familiaris"

    def __init__(self, name, age, breed):
        # Instance attributes (unique to each object)
        self.name  = name
        self.age   = age
        self.breed = breed

    def bark(self):
        return f"{self.name} says: Woof!"

    def description(self):
        return f"{self.name} is a {self.age}-year-old {self.breed}."

# Creating objects (instances)
dog1 = Dog("Buddy", 3, "Golden Retriever")
dog2 = Dog("Max",   5, "German Shepherd")

print(dog1.description())
print(dog2.description())
print(dog1.bark())
print("Species:", dog1.species)   # class attribute, same for all dogs
````

````python
# --- Modifying attributes ---
dog1.age = 4                    # update directly
print(f"{dog1.name} is now {dog1.age} years old.")

# --- Looping over a list of objects ---
dogs = [
    Dog("Luna",   2, "Labrador"),
    Dog("Charlie",6, "Poodle"),
    Dog("Bella",  1, "Bulldog"),
]

print("\nAll dogs:")
for dog in dogs:
    print(" ", dog.description())

# Which dogs are older than 1?
older = [dog.name for dog in dogs if dog.age > 1]
print("\nDogs older than 1 year:", older)
````

````python
# --- Inheritance: a child class inherits from a parent class ---
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound."

class Cat(Animal):                      # Cat inherits from Animal
    def speak(self):                    # override the parent method
        return f"{self.name} says: Meow!"

    def purr(self):
        return f"{self.name} is purring..."

class Duck(Animal):
    def speak(self):
        return f"{self.name} says: Quack!"

animals = [Cat("Whiskers"), Duck("Donald"), Cat("Luna")]

for animal in animals:
    print(animal.speak())               # calls the right version for each type
````

````python
# --- A slightly more practical example: BankAccount ---
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner   = owner
        self.balance = balance
        self.history = []               # list to track transactions

    def deposit(self, amount):
        self.balance += amount
        self.history.append(f"Deposited  ₹{amount}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds!")
            return
        self.balance -= amount
        self.history.append(f"Withdrew   ₹{amount}")

    def statement(self):
        print(f"\n--- Account: {self.owner} ---")
        for entry in self.history:
            print(" ", entry)
        print(f"  Balance: ₹{self.balance}")

acc = BankAccount("Alice", 1000)
acc.deposit(500)
acc.withdraw(200)
acc.withdraw(2000)   # should print "Insufficient funds!"
acc.deposit(100)
acc.statement()
````

### Assignment 1 — Caesar Cipher
A Caesar cipher shifts every letter by a fixed number of positions in the alphabet.
- Only shift letters (a–z, A–Z); leave spaces, digits, and symbols unchanged.
- Wrap around: shifting `'z'` by 1 gives `'a'`.
- Preserve case.

**Example:**
```
caesar("Hello, World!", 3)   →  "Khoor, Zruog!"
caesar("Khoor, Zruog!", -3)  →  "Hello, World!"
```
**Hint:** `ord('a')` gives the ASCII value of `'a'`, and `chr(...)` converts back.

````python
def caesar(text, shift):
    a=ord('a')
    z=ord('z')
    A=ord('A')
    Z=ord('Z')
    ftext=""
    for i in text:
        if(a<=ord(i)<=z):
            ftext+=chr((ord(i)-a+shift)%26 +a)
        elif(A<=ord(i)<=Z):
            ftext+=chr((ord(i)-A+shift)%26 +A)
        else:
            ftext+=i
    return ftext      
        
````

````python
assert caesar("Hello, World!", 3)  == "Khoor, Zruog!", f'Test 1 failed: {caesar("Hello, World!", 3)}'
assert caesar("Khoor, Zruog!", -3) == "Hello, World!", 'Test 2 failed'
assert caesar("xyz", 3)            == "abc",           f'Test 3 (wrap-around): {caesar("xyz", 3)}'
assert caesar("ABC", 1)            == "BCD",           'Test 4 failed'
assert caesar("Hello 123!", 0)     == "Hello 123!",    'Test 5 (zero shift) failed'
print("All Caesar cipher tests passed!")
````

### Assignment 2 — Balanced Brackets
Write `is_balanced(s)` that returns `True` if every opening bracket has a correct matching closing bracket.
Bracket pairs: `()`, `[]`, `{}`.

**Examples:**
```
is_balanced("({[]})")  →  True
is_balanced("([)]")    →  False
is_balanced("")        →  True
```
**Hint:** Use a list as a **stack** — `append` to push, `pop` to pull the last item.

````python
def is_balanced(s):
    pairs={')':'(',']':'[','}':'{'}
    stack=[]
    for char in s:
        if char in "([{":
            stack.append(char)
        elif char in ")]}":
            if not stack:
                return False
            top=stack.pop()
            if pairs[char]!=top:
                return False
    return len(stack)==0
        
````

````python
assert is_balanced("({[]})") == True,  'Test 1 failed'
assert is_balanced("([)]")   == False, 'Test 2 failed'
assert is_balanced("")       == True,  'Test 3 failed'
assert is_balanced("(((")    == False, 'Test 4 failed'
assert is_balanced("{[()]}") == True,  'Test 5 failed'
assert is_balanced("}{")     == False, 'Test 6 failed'
print("All balanced bracket tests passed!")
````

### Assignment 3 — Flatten a Nested List
Write `flatten(nested)` that returns a single flat list regardless of how deeply nested the input is.

**Examples:**
```
flatten([1, [2, 3], [4, [5, 6]]])  →  [1, 2, 3, 4, 5, 6]
flatten([[1, [2]], [[[3]]], 4])    →  [1, 2, 3, 4]
flatten([])                        →  []
```
**Hint:** Use recursion and `isinstance(item, list)`.

````python
def flatten(nested):
    flist=[]
    for item in nested:
        if isinstance(item,list):
            flist.extend(flatten(item))
        else:
            flist.append(item)
    return flist
````

````python
assert flatten([1, [2, 3], [4, [5, 6]]]) == [1, 2, 3, 4, 5, 6]
assert flatten([[1, [2]], [[[3]]], 4])    == [1, 2, 3, 4]
assert flatten([])                        == []
assert flatten([1, 2, 3])                 == [1, 2, 3]
assert flatten([[[[[42]]]]])              == [42]
print("All flatten tests passed!")
````

### Assignment 4 — Run-Length Encoding
Write `encode(s)` and `decode(s)`:
- `encode` compresses consecutive repeated characters: `"aaabbc"` → `"3a2b1c"`
- `decode` reverses it: `"3a2b1c"` → `"aaabbc"`

**Hint for decode:** read digit characters until you hit a letter, then repeat that letter.

````python
def encode(s):
    if not s:
        return ""
    result1=""
    count1=1
    for i in range(1, len(s)):
        if s[i]==s[i-1]:
            count1+=1
        else:
            result1+=str(count1)+s[i-1]
            count1=1
    result1+=str(count1)+s[-1]
    return result1

def decode(s):
    result2=""
    count2=""
    for char in s:
        if char.isdigit():
            count2+=char
        else:
            result2+=char*int(count2)
            count2=""
    return result2
````

````python
assert encode("aaabbbccddddee") == "3a3b2c4d2e"
assert encode("abc")            == "1a1b1c"
assert encode("")               == ""
assert decode("3a3b2c4d2e")     == "aaabbbccddddee"
assert decode("1a1b1c")         == "abc"
assert decode(encode("aaabbccccddddddddde")) == "aaabbccccddddddddde"
print("All run-length encoding tests passed!")
````

### Assignment 5 — Class: Student Report Card
Create a class `Student` with:
- `__init__(self, name)` — store the name and an empty dictionary for subject grades
- `add_grade(self, subject, mark)` — add or update a subject's mark (0–100)
- `average(self)` — return the average mark across all subjects (return `0` if no grades)
- `top_subject(self)` — return the name of the subject with the highest mark
- `report(self)` — print a formatted report card (see expected output below)

**Expected output for `report()`:**
```
Report Card: Alice
  Maths    : 88
  Science  : 73
  English  : 91
Average: 84.00 | Top subject: English
```

````python
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = {}

    def add_grade(self, subject, mark):
        self.grades[subject] = mark

    def average(self):
        if not self.grades:
            return 0
        tmarks = sum(self.grades.values())
        return tmarks / len(self.grades)
        
    def top_subject(self):
        if not self.grades:
            return None
        return max(self.grades, key=self.grades.get)

    def report(self):
        print("Report Card:", self.name)
        for subject in self.grades:
            print(subject, ":", self.grades[subject])
        print("Average:", self.average())
        print("Top subject:", self.top_subject())
````

````python
s = Student("Alice")
s.add_grade("Maths",   88)
s.add_grade("Science", 73)
s.add_grade("English", 91)

assert s.average()     == (88 + 73 + 91) / 3, f'average() failed: {s.average()}'
assert s.top_subject() == "English",           f'top_subject() failed: {s.top_subject()}'

# Edge case: no grades yet
empty = Student("Bob")
assert empty.average() == 0, 'average() on empty student failed'

s.report()   # visually check the output looks right
print("\nAll Student tests passed!")
````

### Assignment 6 (Bonus) — Shapes with Inheritance
Create a base class `Shape` and two child classes `Circle` and `Rectangle`.

- `Shape` has a method `area(self)` that returns `0` by default.
- `Circle(radius)` overrides `area()` to return πr² (use `math.pi`).
- `Rectangle(width, height)` overrides `area()` to return width × height.
- Both child classes also have a `describe(self)` method that prints something like:
  - `"Circle with radius 5, area = 78.54"`
  - `"Rectangle 4x6, area = 24"`

**Hint:** Call the child's own `area()` inside `describe()`.

````python
import math

class Shape:
    def area(self):
        return 0

class Circle(Shape):
    def __init__(self, radius):
        self.radius=radius

    def area(self):
        return math.pi*(self.radius*self.radius)

    def describe(self):
        print("Circle with radius", self.radius)
        print("area:", self.area())

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width=width
        self.height=height

    def area(self):
        return self.width*self.height

    def describe(self):
        print("Rectangle", self.width,"x",self.height)
        print("area:", self.area())
````

````python
import math

c = Circle(5)
r = Rectangle(4, 6)

assert abs(c.area() - math.pi * 25) < 1e-9, f'Circle area failed: {c.area()}'
assert r.area() == 24,                        f'Rectangle area failed: {r.area()}'

shapes = [Circle(3), Rectangle(2, 5), Circle(7)]
for shape in shapes:
    shape.describe()   # visually check each line

print("\nAll Shape tests passed!")
````
